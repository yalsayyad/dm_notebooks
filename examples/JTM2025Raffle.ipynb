```python
def weighted_sample_without_replacement(items, k):
    items = items.copy()
    winners = []

    for _ in range(min(k, len(items))):
        winner = random.choices(
            population=list(items.keys()),
            weights=list(items.values()),
            k=1
        )[0]

        winners.append(winner)
        del items[winner]  # remove winner (no replacement)

    return winners, items
```
